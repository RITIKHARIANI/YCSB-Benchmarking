```python
#Workload A
import matplotlib.pyplot as plt

#Data Processing
f = open("A.txt","r")
data = f.read()
data = data.split('\n')

headings = data[0].split(',')
record_count=[]
read_latency = []
update_latency = []
throughput = []
for i in range(1,6):
    record_count.append(data[i].split(',')[0])
    read_latency.append(data[i].split(',')[1])
    update_latency.append(data[i].split(',')[2])
    throughput.append(data[i].split(',')[3])
f.close()

record_count = list(map(float,record_count))
read_latency = list(map(float,read_latency))
update_latency = list(map(float,update_latency))
throughput = list(map(float,throughput))

#Plotting
plt.figure(1)
plt.plot(record_count,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(record_count,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload A")
plt.xlabel(headings[0])
plt.ylabel("Average Latency (us)")
plt.show()
plt.figure(2)
plt.plot(record_count,throughput,label="Throughput",marker="o",color="red")
plt.legend()
plt.title("Workload A")
plt.xlabel(headings[0])
plt.ylabel(headings[3])
plt.show()
```

```python
#Workload F
import matplotlib.pyplot as plt

#Data Processing
f = open("F.txt","r")
data = f.read()
data = data.split('\n')

headings = data[0].split(',')
record_count=[]
read_latency = []
read_modify_write_latency = []
update_latency = []
throughput = []
for i in range(1,6):
    record_count.append(data[i].split(',')[0])
    read_latency.append(data[i].split(',')[1])
    read_modify_write_latency.append(data[i].split(',')[2])
    update_latency.append(data[i].split(',')[3])
    throughput.append(data[i].split(',')[4])
f.close()

record_count = list(map(float,record_count))
read_latency = list(map(float,read_latency))
read_modify_write_latency = list(map(float,read_modify_write_latency))
update_latency = list(map(float,update_latency))
throughput = list(map(float,throughput))

#Plotting

plt.figure(1)
plt.plot(record_count,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(record_count,read_modify_write_latency,label="Avg Read_Modify_Write_Latency",marker="o")
plt.plot(record_count,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload F")
plt.xlabel(headings[0])
plt.ylabel("Average Latency (us)")
plt.show()
plt.figure(2)
plt.plot(record_count,throughput,label="Throughput",marker="o",color="red")
plt.legend()
plt.title("Workload F")
plt.xlabel(headings[0])
plt.ylabel(headings[4])
plt.show()
```
